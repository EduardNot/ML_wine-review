
### H2O Gradient Boosting Machine

```python
#!pip install h2o
import h2o
from h2o.estimators import H2OGradientBoostingEstimator
h2o.init()
```

```python
gbm_data = h2o.import_file('wine_reviews_for_gbm.csv')
```

```python
gbm_data.head()
```

```python
train, val = gbm_data.split_frame(
    ratios=[0.8],
    seed=1234)
```

```python
#Here we use previously chosen categorical features to predict wine price.

predictors = categorical_columns
response = 'price'
```

```python
gbm = H2OGradientBoostingEstimator()
gbm.train(x=predictors, y=response, training_frame=train)
```

```python
performance = gbm.model_performance(val)
performance
```

```python
gbm.feature_interaction()
```

```python
#Variable importance

gbm.varimp(use_pandas=True)
```
